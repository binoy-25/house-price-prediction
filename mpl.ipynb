```python
import matplotlib.pyplot as plt
x=[1,2,3]
y1=[10,20,30]
y2=[15,25,35]
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple line chart")
plt.show()
```

```python
x=['a','b','c']
y=[10,20,15]
plt.bar(x,y)
plt.show()
```

```python
data=[1,2,2,3,3,4]
plt.hist(data)
plt.show()
```

```python
x=[1,2,3,4]
y=[10,15,25,35]
plt.scatter(x,y)
plt.show()
```

```python
x=['a','b','c']
y=[30,40,30]
plt.pie(y,labels=x)
plt.show()
```

```python
plt.plot(x,y,linestyle="--",marker="o")
plt.grid()
plt.show()
```
